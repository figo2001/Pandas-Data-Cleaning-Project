## String Replace

```python
df.head(4)
```

```python
df['Salary']=df['Salary'].str.replace('£', '').str.replace('K','000')
```

```python
df.head(4)
```

```python
df['Salary']=df['Salary'].str.replace(r'\(Employer est.\)', '', regex=True)
df['Salary']=df['Salary'].str.replace(r'\(Glassdoor est.\)', '', regex=True)
```

```python
df.head()
```

```python
import re
```

```python
def convert_to_yearly(salary):
    if 'Per Hour' in str(salary):
        hourly_rate=re.findall(r'\d+\.?\d*', salary)

        if hourly_rate:
            
            return float(hourly_rate[0])*2080  # Convert to yearly salary
    return salary
```

```python
df['Salary']=df['Salary'].apply(convert_to_yearly)
```

```python
df.head(4)
```

## Split Column

```python
df['Salary']=df['Salary'].astype(str)
```

```python
df[['salary_min', 'salary_max']]=df['Salary'].str.split('-', expand=True)
```

```python
df.head(4)
```

## Remove Whitespaces

```python
df['salary_min']=df['salary_min'].str.strip()
df['salary_max']=df['salary_max'].str.strip()
```

```python
df['salary_max'].dtype
```

## Now Converting the columns to numeric

**We are using coerce here cause in our dataset there is null values and if we ignore this, then we should face some issues on type convertion.**

```python
df['salary_min']=pd.to_numeric(df['salary_min'], errors='coerce')
df['salary_max']=pd.to_numeric(df['salary_max'], errors='coerce')
```

```python
df['salary_max'].dtype
```

#### Converting Our Location column into string

```python
df['Location']=df['Location'].astype(str)
```

```python
df[['city', 'country']]=df['Location'].str.split(',', n=1, expand=True)
```

```python
df.head(4)
```

### Apply Skills Column to One Hot Encoding

So Python will create columns for each possible skill in the skills column and we will get a binary result

```python
skills_split=df['Skills'].str.get_dummies(sep=', ')
```

```python
skills_split
```

### Now combining two dataframes

```python
df_with_skills=pd.concat([df, skills_split], axis=1)
```

```python
df_with_skills.head()
```

## Data Visualization

### i) Word Cloud

```python
from wordcloud import WordCloud
```

```python
skill_frequencies=df_with_skills.iloc[:, df_with_skills.columns.get_loc('AI'):].sum()

skill_frequencies
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wordcloud=WordCloud(width=1000, height=700, background_color='white').generate_from_frequencies(skill_frequencies)

plt.imshow(wordcloud, interpolation='bilinear')

plt.show()
```

### ii) Bar Chart for Top 10 Skills 

```python
skill_columns=df_with_skills.iloc[:, df_with_skills.columns.get_loc('AI'):].sum().sort_values(ascending=False)

skill_columns
```

```python
top_10_skills=skill_columns.head(10)

top_10_skills
```

**plt.gca().invert_yaxis() --> This will help to see the chart from descending order to ascending order only when we use barh plot in pandas.**

```python
top_10_skills.plot(kind='barh', figsize=(10,6), color='green')
plt.xticks(rotation=45);

plt.title('Top 10 Most Popular Skills', fontsize=16)

plt.gca().invert_yaxis()
plt.show()
```
